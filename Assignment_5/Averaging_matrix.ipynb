```python
import numpy as np
from itertools import product

def mat(n,k):
    A=np.zeros((n,n))

    for i,j,t in product(range(n),range(n),range(k+1)):
        if abs(i-j)==t:
            A[i][j]=1 

    return A
```

```python
print(mat(6,2))
```

```python
def avg(n,k):    # creating the average matrix operator
    
    A=mat(n,k)

    for i in range(n):
        r = np.sum(A[i])
        A[i]=A[i]/r

    return A
```

```python
print(avg(6,2))
```

```python
n=20
k=2

x = np.array([(-1)**j for j in range(n)])
x
```

```python
A=avg(n,k)

A@x
```

```python
def f(x):
    return x**2 + 2*np.sin(10*x)
```

```python
n=100
k=10

x = np.linspace(0, 10, n)

avg(n,k)@f(x)
```

```python
import matplotlib.pyplot as plt

n=20

x = np.array([(-1)**j for j in range(n)])

for k in range(0,25,3):
    y=avg(n,k)@x

    plt.plot(x,y)

    plt.title("Averaging x")
    plt.xlabel('x')
    plt.ylabel('Avg x')

    plt.show()
```

Averaging out the vector as k increases

```python
n=1000

x = np.linspace(0, 10, n)

for k in range(0,35,5):
    y=avg(n,k)@f(x)

    plt.plot(x,y)

    plt.title("Averaging f(x)")
    plt.xlabel('x')
    plt.ylabel('Avg f(x)')
    
    plt.show()
```

As k increases f(x) smooths out i.e it averages with nearby values at each inputs. Note: at k=0 it is exactly the x vs f(x) graph.
