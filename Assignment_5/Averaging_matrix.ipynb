```python
import numpy as np
from itertools import product

def mat(n,k):
    A=np.zeros((n,n))

    for i,j,t in product(range(n),range(n),range(k+1)):
        if abs(i-j)==t:
            A[i][j]=1 

    return A
```

```python
print(mat(6,2))
```

```python
def avg(n,k):    # creating the average matrix operator
    
    A=mat(n,k)

    for i in range(n):
        r = np.sum(A[i])
        A[i]=A[i]/r

    return A
```

```python
print(avg(6,2))
```
