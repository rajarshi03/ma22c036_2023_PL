```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    t=x   # setting temp variable for ease on eyes
    y=np.zeros_like(x)  # creating array for storing results


    y[t<0]=x[t<0]*np.sin(x[t<0]**2)   # first component of piecewise function

    y[t>=0]=x[t>=0]*np.cos(x[t>=0])   # first component of piecewise function


    return y
```

```python
x=np.linspace(-5,5,1000001)
```

```python
%timeit y=f(x)
```

```python
y=f(x)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("Plot of the Piecewise Function")

plt.show()
```
