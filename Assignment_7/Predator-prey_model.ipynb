```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn; seaborn.set();

from scipy.integrate import solve_ivp
import numpy as np
```

```python
def f(t, y, alpha=1.1, beta=0.4, delta=0.1, gamma=0.1):
    r, x = y

    return np.array([alpha*r - beta*r*x,
                    delta*r*x - gamma*x])
```

```python
sol = solve_ivp(f, [0, 70], [5,2], rtol=1e-6)
```

```python
fig = plt.figure(); ax = fig.gca()

curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['Rabbits(r)','Foxes(x)']);

ax.set_title('Predator-prey model for rabbits and foxes')
```

```python
# for randomly chosen initial values r(0) and x(0) between 1 and 9
import random

for i in range(10):
    r0=random.uniform(1,9)
    x0=random.uniform(1,9)

    sol = solve_ivp(f, [0, 70], [r0,x0], rtol=1e-6)


    fig = plt.figure(); ax = fig.gca()

    curves = ax.plot(sol.t, sol.y.T)
    ax.legend(curves, ['Rabbits(r)','Foxes(x)']);

    ax.set_title(f'r(0)={r0} and x(0)={x0}')
```
