```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn; seaborn.set();

from scipy.integrate import solve_ivp
import numpy as np
```

```python
def f(t, y, alpha=1.1, beta=0.4, delta=0.1, gamma=0.1):
    r, x = y

    return np.array([alpha*r - beta*r*x,
                    delta*r*x - gamma*x])
```

```python
sol = solve_ivp(f, [0, 70], [5,2], rtol=1e-6)
```

```python
fig = plt.figure(); ax = fig.gca()

curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['Rabbits(r)','Foxes(x)']);

ax.set_title('Predator-prey model for rabbits and foxes')
```

Taking derivatives 0, we get two equilibria for the system: $r=0,x=0$ and $r=\frac{\gamma}{\delta}, x=\frac{\alpha}{\beta}$

```python
# for randomly chosen initial values r(0) and x(0) between 1 and 9
import random

plt.figure(figsize=(8,8))

for i in range(10):
    r0=random.uniform(1,9)
    x0=random.uniform(1,9)

    sol = solve_ivp(f, [0, 70], [r0,x0], rtol=1e-6)


    plt.plot(sol.y[0],sol.y[1], label=f'r(0)={r0:.2f}, x(0)={x0:.2f}')


# equilibria

alpha=1.1; beta=0.4; delta=0.1; gamma=0.1;

plt.scatter([0,gamma/delta],[0,alpha/beta],label='equilibria')


plt.legend()
plt.xlabel("Rabbit")
plt.ylabel('Fox')
plt.title('Phase plot: Rabbit vs Fox')
plt.show()
```
