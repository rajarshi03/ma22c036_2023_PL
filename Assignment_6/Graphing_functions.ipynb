```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,10,1000)  #getting 1000 points in [0,10] evenly

y = np.sin(x)       # getting value of those points under sin(x)
```

```python
# ploting the function

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("sin(x)")

plt.title("Plot of sin(x) on [0,10]")

plt.show()
```

```python
x = np.linspace(-5,5,80)
y = np.linspace(-5,5,80)

X, Y = np.meshgrid(x, y)

z = np.cos(np.sqrt(X**2+Y**2))
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, rstride=1, cstride=1,cmap='viridis', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
